```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json

```

```python
df_superhero_info = pd.read_csv('superhero_info - superhero_info.csv')
df_superhero_info.info()
df_superhero_info.head()
```

```python
df_superhero_powers = pd.read_csv('superhero_powers - superhero_powers.csv')
df_superhero_powers.info()
df_superhero_powers.head(5)
```

```python
## example merge with our 2 dataframes
df_superhero_info = pd.concat((df_superhero_info,df_superhero_powers['Powers']),axis=1)
df_superhero_info.head(5)
```

```python
## save the 2 new columns into the dataframe
df_superhero_info[['Hero','Publisher']] =df_superhero_info['Hero|Publisher'].str.split('|',expand=True)
## drop the original column 
df_superhero_info = df_superhero_info.drop(columns=['Hero|Publisher'])
df_superhero_info.head(2)
```

```python
## save the 2 new columns into the dataframe
df_superhero_info[['Height','Weight']] =df_superhero_info['Measurements'].str.split(',',expand=True)
df_superhero_info = df_superhero_info.drop(columns=['Measurements'])

# Make a list of all characters to replace
to_replace_height = ['{',"'",'Height',"'",':',"'"," cm"]
# run a loop to replace all of the characters in the list at once
for char in to_replace_height:
    df_superhero_info['Height'] = df_superhero_info['Height'].str.replace(char,'',regex=False).str.strip()


to_replace_weight = ['{',"'",'Weight',"'",':',"'",'}'," kg"]
# run a loop to replace all of the characters in the list at once
for char in to_replace_weight:
    df_superhero_info['Weight'] = df_superhero_info['Weight'].str.replace(char,'',regex=False).str.strip()

df_superhero_info.head()
```

```python
cols_to_make = df_superhero_info['Powers'].str.split(',').explode().unique().tolist()
cols_to_make
```

```python
for col in cols_to_make:
    df_superhero_info[col] = df_superhero_info['Powers'].str.contains(col)
#df_superhero_info = df_superhero_info.drop(columns=['Powers'])
df_superhero_info.head()
```

```python
df_superhero_info = df_superhero_info.drop(columns=['Powers'])
df_superhero_info.head()
```

```python
df_superhero_info.dropna(subset=['Weight'], inplace=True)
```

```python
df_superhero_info['Weight'] = df_superhero_info['Weight'].astype(float)
```

```python
mean_Weight = df_superhero_info.groupby('Super Speed')['Weight'].mean()
mean_Weight
```

```python
df_superhero_info.dropna(subset=['Height'], inplace=True)
```

```python
df_superhero_info['Height'] = df_superhero_info['Height'].astype(float)
```
